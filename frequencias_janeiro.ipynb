# CÃ¡lculos de frequencias

`````python
# Calcular a frequÃªncia absoluta
freq_abs = janeiro_filtrado['setor'].value_counts()

# Calcular a frequÃªncia relativa
freq_rel = janeiro_filtrado['setor'].value_counts(normalize=True)

# Calcular a frequÃªncia relativa acumulada
freq_rel_acum = freq_rel.cumsum()

# Calcular o percentual
percentual = freq_rel * 100

# Calcular o percentual acumulado
percentual_acum = freq_rel_acum * 100

# Criar um DataFrame para exibir os resultados
frequencia_df = pd.DataFrame({
    'FrequÃªncia Absoluta': freq_abs,
    'FrequÃªncia Relativa': freq_rel,
    'FrequÃªncia Relativa Acumulada': freq_rel_acum,
    'Percentual (%)': percentual,
    'Percentual Acumulado (%)': percentual_acum
})

display(frequencia_df.round(2))
`````

## CÃ¡lculos de frequencias // explicaÃ§Ã£o

Este cÃ³digo calcula e exibe as estatÃ­sticas de frequÃªncia para a coluna `'setor'` no seu DataFrame filtrado (`janeiro_filtrado`).

* **`freq_abs = janeiro_filtrado['setor'].value_counts()`**: Calcula a **frequÃªncia absoluta**, que Ã© uma contagem simples de quantas vezes cada valor Ãºnico aparece na coluna `'setor'`.

* **`freq_rel = janeiro_filtrado['setor'].value_counts(normalize=True)`**: Calcula a **frequÃªncia relativa**. O argumento `normalize=True` instrui o pandas a dividir a contagem de cada setor pelo nÃºmero total de registros, resultando na proporÃ§Ã£o de cada um.

* **`freq_rel_acum = freq_rel.cumsum()`**: Calcula a **frequÃªncia relativa acumulada**. O mÃ©todo `cumsum()` calcula a soma cumulativa das frequÃªncias relativas, permitindo que vocÃª veja o total acumulado.

* **`percentual = freq_rel * 100`**: Calcula o **percentual** multiplicando a frequÃªncia relativa de cada setor por 100.

* **`percentual_acum = freq_rel_acum * 100`**: Calcula o **percentual acumulado** multiplicando a frequÃªncia relativa acumulada por 100.

* **`frequencia_df = pd.DataFrame({...})`**: Cria um novo DataFrame chamado `frequencia_df` para exibir de forma organizada todas as frequÃªncias e percentuais calculados.

* **`display(frequencia_df)`**: Exibe o DataFrame `frequencia_df`, mostrando os resultados da sua anÃ¡lise.

# CÃ¡lculo da mÃ©dia de pacientes que chegam por hora, por mÃªs

`````python
# CÃ¡lculo da mÃ©dia de pacientes que chegam por hora, por mÃªs


import pandas as pd

# Supondo que janeiro_filtrado jÃ¡ seja um DataFrame filtrado corretamente
# Fazendo uma cÃ³pia explÃ­cita para evitar os avisos de alteraÃ§Ã£o em slice
janeiro_filtrado = janeiro_filtrado.copy()

# Convertendo a coluna 'data_chegada' para datetime
janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')

# Extraindo a hora da chegada
janeiro_filtrado['hora_chegada'] = janeiro_filtrado['data_chegada'].dt.hour

# Definindo o ciclo 07hâ06h
horas_ciclo = list(range(7, 24)) + list(range(0, 7))
janeiro_filtrado['hora_chegada'] = pd.Categorical(janeiro_filtrado['hora_chegada'],
                                                  categories=horas_ciclo,
                                                  ordered=True)

# Agrupando por hora e contando pacientes, com observed=False explÃ­cito para evitar warnings futuros
pacientes_por_hora = janeiro_filtrado.groupby('hora_chegada', observed=False).size()

# Calculando a mÃ©dia de pacientes por hora
media_pacientes_por_hora = pacientes_por_hora.mean()

print(f"MÃ©dia de pacientes por hora (ciclo de 07h-06h): {media_pacientes_por_hora:.2f}")
`````

# MÃ©dia de pacientes que chegam por hora, por dia, em um mÃªs (janeiro)

`````python
# Certificando-se de que 'data_chegada' Ã© do tipo datetime
janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')

# Extraindo a data e a hora para colunas separadas
janeiro_filtrado['dia'] = janeiro_filtrado['data_chegada'].dt.date
janeiro_filtrado['hora'] = janeiro_filtrado['data_chegada'].dt.hour

# Agrupando por dia e hora e contando o nÃºmero de pacientes
pacientes_por_dia_hora = janeiro_filtrado.groupby(['dia', 'hora']).size()

# Calculando a mÃ©dia de pacientes por hora para todos os dias
media_por_hora_dia = pacientes_por_dia_hora.groupby('hora').mean()

print("MÃ©dia de pacientes por hora por dia:")
display(media_por_hora_dia.round(2))
`````

## MÃ©dia de pacientes que chegam por hora, por dia, em um mÃªs (janeiro) // explicaÃ§Ã£o


* `janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')`: Esta linha garante que a coluna `'data_chegada'` esteja em um formato de data e hora adequado, o que Ã© necessÃ¡rio para as etapas seguintes.

* `janeiro_filtrado['dia'] = janeiro_filtrado['data_chegada'].dt.date`: Esta linha extrai apenas a parte da data (ano, mÃªs, dia) da coluna `'data_chegada'` e a insere em uma nova coluna chamada `'dia'`.

* `janeiro_filtrado['hora'] = janeiro_filtrado['data_chegada'].dt.hour`: Esta linha extrai apenas a hora da coluna `'data_chegada'` e cria uma nova coluna chamada `'hora'`.

* `pacientes_por_dia_hora = janeiro_filtrado.groupby(['dia', 'hora']).size()`: Este Ã© um passo fundamental. Ele agrupa o DataFrame tanto pela coluna `'dia'` quanto pela `'hora'` e, em seguida, conta o nÃºmero de pacientes (`.size()`) em cada um desses grupos. O resultado Ã© uma sÃ©rie que mostra o nÃºmero de pacientes para cada hora de cada dia.

* `media_por_hora_dia = pacientes_por_dia_hora.groupby('hora').mean()`: Esta linha pega o resultado da etapa anterior e o agrupa novamente, desta vez apenas pela `'hora'`. Em seguida, para cada hora, calcula a mÃ©dia (`.mean()`) das contagens de pacientes de todos os dias.

* `print(...)` e `display(...)`: Estas linhas simplesmente imprimem um tÃ­tulo e depois exibem o resultado final, que Ã© a mÃ©dia de pacientes para cada hora do dia, arredondada para duas casas decimais.

# Reproduzindo a anÃ¡lise anterior nos outros meses

Explicar como reproduzir eficientemente toda a anÃ¡lise (carregamento dos dados, filtragem, agregaÃ§Ã£o e visualizaÃ§Ã£o com grÃ¡fico de barras empilhadas) para mais cinco meses, partindo da anÃ¡lise inicial feita com dados de um Ãºnico mÃªs em arquivo Excel.

## SoluÃ§Ã£o proposta

Criar uma funÃ§Ã£o Python chamada analyze_monthly_data que:
	1.	Recebe o caminho do arquivo Excel como parÃ¢metro;
	2.	Carrega os dados;
	3.	Aplica os filtros necessÃ¡rios;
	4.	Realiza a agregaÃ§Ã£o dos dados;
	5.	Gera um grÃ¡fico de barras empilhadas com os resultados.

Com isso, basta chamar a funÃ§Ã£o para cada novo arquivo mensal e repetir toda a anÃ¡lise automaticamente.

`````python
def analyze_monthly_data(file_path):
    """
    Analyzes monthly patient data from an Excel file, generating a stacked bar chart
    of average patients per hour by sector.

    Args:
        file_path (str): The path to the Excel file.
    """
    # Extrair o nome do mÃªs do caminho do arquivo para o tÃ­tulo do grÃ¡fico
    month_name = file_path.split('.')[0].capitalize()

    # Carregar os dados do arquivo Excel
    df = pd.read_excel(file_path, usecols=[0, 1, 5, 18, 20, 22, 28, 29, 30, 31, 32, 33, 34, 35, 37, 44])

    # Filtrar o DataFrame para incluir apenas os setores de interesse
    setores_de_interesse = ['CONSULTORIO CLINICO', 'CONSULTORIO PEDIATRICO', 'SALA DE SUTURA']
    df_filtrado = df[df['setor'].isin(setores_de_interesse)].copy()

    # Converter 'data_chegada' para datetime
    df_filtrado['data_chegada'] = pd.to_datetime(df_filtrado['data_chegada'], errors='coerce')

    # Calcular o total de pacientes por dia e hora para cada setor
    pacientes_por_dia_hora_setor = df_filtrado.groupby([df_filtrado['data_chegada'].dt.date, df_filtrado['data_chegada'].dt.hour, 'setor']).size().unstack(fill_value=0)

    # Calcular a mÃ©dia de pacientes por hora para cada setor
    media_por_hora_setor = pacientes_por_dia_hora_setor.groupby(level=1).mean()

    # Criar o grÃ¡fico de barras empilhadas
    ax = media_por_hora_setor.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='viridis')

    # Adicionar os valores em cada segmento da barra
    for c in ax.containers:
        labels = [f'{v.get_height():.2f}' if v.get_height() > 0 else '' for v in c]
        ax.bar_label(c, labels=labels, label_type='center')

    # Adicionar tÃ­tulos e rÃ³tulos
    plt.title(f'MÃ©dia de Pacientes por Hora do Dia por Especialidade - {month_name}')
    plt.xlabel('Hora do Dia')
    plt.ylabel('MÃ©dia de Pacientes')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--')
    plt.legend(title='Setor')

    # Exibir o grÃ¡fico
    plt.show()
`````

# Juntando todas os meses em um Ãºnico dataframe

Usando os dados dos seguintes arquivos: âdados_jan.xlsxâ, âdados_fev.xlsxâ, âdados_mar.xlsxâ, âdados_abr.xlsxâ, âdados_maio.xlsxâ e âdados_jun.xlsxâ, combine todos os dados mensais em um Ãºnico DataFrame para realizar uma anÃ¡lise de tendÃªncia ao longo de todo o perÃ­odo, identificando padrÃµes sazonais ou mudanÃ§as de longo prazo no volume de pacientes.

Carregue os dados de cada um dos arquivos mensais do Excel e combine-os em um Ãºnico DataFrame abrangente. **TambÃ©m adicionarei uma coluna chamada 'mÃªs' para identificar a origem dos dados.**

Agora vou implementar os passos para carregar e combinar os dados dos arquivos Excel especificados. Isso envolve **criar uma lista com os caminhos dos arquivos, iterar sobre eles para carregar cada arquivo em um DataFrame, adicionar uma coluna 'mÃªs' a cada DataFrame e, em seguida, concatenÃ¡-los em um Ãºnico DataFrame.**

`````python
# Juntando todas os meses em um Ãºnico dataframe

import pandas as pd

# 1. Create a list of file paths
file_paths = ["janeiro25.xlsx", "fevereiro25.xlsx", "marco25.xlsx", "abril25.xlsx", "maio25.xlsx", "junho25.xlsx"]

# 2. Initialize an empty list to store DataFrames
all_months_data = []

# 3. Loop through each file path
for file_path in file_paths:
    # a. Load the Excel file
    df = pd.read_excel(file_path, usecols=[0, 1, 5, 18, 20, 22, 28, 29, 30, 31, 32, 33, 34, 35, 37, 44])

    # b. Extract the month name
    month_name = file_path.split('.')[0][:-2]

    # c. Add a 'month' column
    df['month'] = month_name

    # d. Append the DataFrame to the list
    all_months_data.append(df)

# 4. Combine all DataFrames
combined_df = pd.concat(all_months_data, ignore_index=True)

# 5. Display the first few rows of the combined DataFrame
display(combined_df.head())
`````

# MÃ©dia de pacientes novos, por hora, por especialidade, em um dia

`````python
# CONSOLIDADO: mÃ©dia de chegadas, por hora, por especialidade

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Filter for the relevant sectors from the combined dataframe
setores_de_interesse = ['CONSULTORIO CLINICO', 'CONSULTORIO PEDIATRICO', 'SALA DE SUTURA']
df_filtrado_combined = combined_df[combined_df['setor'].isin(setores_de_interesse)].copy()

# Ensure 'data_chegada' is in datetime format
df_filtrado_combined['data_chegada'] = pd.to_datetime(df_filtrado_combined['data_chegada'], errors='coerce')


# 2. Calculate the total number of patients per day and hour for each specialty
pacientes_por_dia_hora_setor = df_filtrado_combined.groupby([df_filtrado_combined['data_chegada'].dt.date, df_filtrado_combined['data_chegada'].dt.hour, 'setor']).size().unstack(fill_value=0)

# 3. Calculate the average number of patients per hour for each specialty
media_por_hora_setor = pacientes_por_dia_hora_setor.groupby(level=1).mean()

# 4. Create the stacked bar chart with a new color palette
ax = media_por_hora_setor.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='Set2')

# 5. Add the values to each segment of the bar
for c in ax.containers:
    labels = [f'{v.get_height():.2f}' if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center')

# 6. Add titles and labels
plt.title('MÃ©dia de Pacientes por Hora do Dia por Especialidade (Todos os Meses)')
plt.xlabel('Hora do Dia')
plt.ylabel('MÃ©dia de Pacientes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.legend(title='Setor')

# Display the chart
plt.show()
`````
