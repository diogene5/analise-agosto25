# Cálculos de frequencias

`````python
# Calcular a frequência absoluta
freq_abs = janeiro_filtrado['setor'].value_counts()

# Calcular a frequência relativa
freq_rel = janeiro_filtrado['setor'].value_counts(normalize=True)

# Calcular a frequência relativa acumulada
freq_rel_acum = freq_rel.cumsum()

# Calcular o percentual
percentual = freq_rel * 100

# Calcular o percentual acumulado
percentual_acum = freq_rel_acum * 100

# Criar um DataFrame para exibir os resultados
frequencia_df = pd.DataFrame({
    'Frequência Absoluta': freq_abs,
    'Frequência Relativa': freq_rel,
    'Frequência Relativa Acumulada': freq_rel_acum,
    'Percentual (%)': percentual,
    'Percentual Acumulado (%)': percentual_acum
})

display(frequencia_df.round(2))
`````

## Cálculos de frequencias // explicação

Este código calcula e exibe as estatísticas de frequência para a coluna `'setor'` no seu DataFrame filtrado (`janeiro_filtrado`).

* **`freq_abs = janeiro_filtrado['setor'].value_counts()`**: Calcula a **frequência absoluta**, que é uma contagem simples de quantas vezes cada valor único aparece na coluna `'setor'`.

* **`freq_rel = janeiro_filtrado['setor'].value_counts(normalize=True)`**: Calcula a **frequência relativa**. O argumento `normalize=True` instrui o pandas a dividir a contagem de cada setor pelo número total de registros, resultando na proporção de cada um.

* **`freq_rel_acum = freq_rel.cumsum()`**: Calcula a **frequência relativa acumulada**. O método `cumsum()` calcula a soma cumulativa das frequências relativas, permitindo que você veja o total acumulado.

* **`percentual = freq_rel * 100`**: Calcula o **percentual** multiplicando a frequência relativa de cada setor por 100.

* **`percentual_acum = freq_rel_acum * 100`**: Calcula o **percentual acumulado** multiplicando a frequência relativa acumulada por 100.

* **`frequencia_df = pd.DataFrame({...})`**: Cria um novo DataFrame chamado `frequencia_df` para exibir de forma organizada todas as frequências e percentuais calculados.

* **`display(frequencia_df)`**: Exibe o DataFrame `frequencia_df`, mostrando os resultados da sua análise.

# Cálculo da média de pacientes que chegam por hora, por mês

`````python
# Cálculo da média de pacientes que chegam por hora, por mês


import pandas as pd

# Supondo que janeiro_filtrado já seja um DataFrame filtrado corretamente
# Fazendo uma cópia explícita para evitar os avisos de alteração em slice
janeiro_filtrado = janeiro_filtrado.copy()

# Convertendo a coluna 'data_chegada' para datetime
janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')

# Extraindo a hora da chegada
janeiro_filtrado['hora_chegada'] = janeiro_filtrado['data_chegada'].dt.hour

# Definindo o ciclo 07h–06h
horas_ciclo = list(range(7, 24)) + list(range(0, 7))
janeiro_filtrado['hora_chegada'] = pd.Categorical(janeiro_filtrado['hora_chegada'],
                                                  categories=horas_ciclo,
                                                  ordered=True)

# Agrupando por hora e contando pacientes, com observed=False explícito para evitar warnings futuros
pacientes_por_hora = janeiro_filtrado.groupby('hora_chegada', observed=False).size()

# Calculando a média de pacientes por hora
media_pacientes_por_hora = pacientes_por_hora.mean()

print(f"Média de pacientes por hora (ciclo de 07h-06h): {media_pacientes_por_hora:.2f}")
`````

`````python
# Certificando-se de que 'data_chegada' é do tipo datetime
janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')

# Extraindo a data e a hora para colunas separadas
janeiro_filtrado['dia'] = janeiro_filtrado['data_chegada'].dt.date
janeiro_filtrado['hora'] = janeiro_filtrado['data_chegada'].dt.hour

# Agrupando por dia e hora e contando o número de pacientes
pacientes_por_dia_hora = janeiro_filtrado.groupby(['dia', 'hora']).size()

# Calculando a média de pacientes por hora para todos os dias
media_por_hora_dia = pacientes_por_dia_hora.groupby('hora').mean()

print("Média de pacientes por hora por dia:")
display(media_por_hora_dia.round(2))
`````

## Média de pacientes que chegam por hora, por dia, em um mês (janeiro) // explicação


* `janeiro_filtrado['data_chegada'] = pd.to_datetime(janeiro_filtrado['data_chegada'], errors='coerce')`: Esta linha garante que a coluna `'data_chegada'` esteja em um formato de data e hora adequado, o que é necessário para as etapas seguintes.

* `janeiro_filtrado['dia'] = janeiro_filtrado['data_chegada'].dt.date`: Esta linha extrai apenas a parte da data (ano, mês, dia) da coluna `'data_chegada'` e a insere em uma nova coluna chamada `'dia'`.

* `janeiro_filtrado['hora'] = janeiro_filtrado['data_chegada'].dt.hour`: Esta linha extrai apenas a hora da coluna `'data_chegada'` e cria uma nova coluna chamada `'hora'`.

* `pacientes_por_dia_hora = janeiro_filtrado.groupby(['dia', 'hora']).size()`: Este é um passo fundamental. Ele agrupa o DataFrame tanto pela coluna `'dia'` quanto pela `'hora'` e, em seguida, conta o número de pacientes (`.size()`) em cada um desses grupos. O resultado é uma série que mostra o número de pacientes para cada hora de cada dia.

* `media_por_hora_dia = pacientes_por_dia_hora.groupby('hora').mean()`: Esta linha pega o resultado da etapa anterior e o agrupa novamente, desta vez apenas pela `'hora'`. Em seguida, para cada hora, calcula a média (`.mean()`) das contagens de pacientes de todos os dias.

* `print(...)` e `display(...)`: Estas linhas simplesmente imprimem um título e depois exibem o resultado final, que é a média de pacientes para cada hora do dia, arredondada para duas casas decimais.

# Reproduzindo a análise anterior nos outros meses

Explicar como reproduzir eficientemente toda a análise (carregamento dos dados, filtragem, agregação e visualização com gráfico de barras empilhadas) para mais cinco meses, partindo da análise inicial feita com dados de um único mês em arquivo Excel.

## Solução proposta

Criar uma função Python chamada analyze_monthly_data que:
	1.	Recebe o caminho do arquivo Excel como parâmetro;
	2.	Carrega os dados;
	3.	Aplica os filtros necessários;
	4.	Realiza a agregação dos dados;
	5.	Gera um gráfico de barras empilhadas com os resultados.

Com isso, basta chamar a função para cada novo arquivo mensal e repetir toda a análise automaticamente.

`````python
def analyze_monthly_data(file_path):
    """
    Analyzes monthly patient data from an Excel file, generating a stacked bar chart
    of average patients per hour by sector.

    Args:
        file_path (str): The path to the Excel file.
    """
    # Extrair o nome do mês do caminho do arquivo para o título do gráfico
    month_name = file_path.split('.')[0].capitalize()

    # Carregar os dados do arquivo Excel
    df = pd.read_excel(file_path, usecols=[0, 1, 5, 18, 20, 22, 28, 29, 30, 31, 32, 33, 34, 35, 37, 44])

    # Filtrar o DataFrame para incluir apenas os setores de interesse
    setores_de_interesse = ['CONSULTORIO CLINICO', 'CONSULTORIO PEDIATRICO', 'SALA DE SUTURA']
    df_filtrado = df[df['setor'].isin(setores_de_interesse)].copy()

    # Converter 'data_chegada' para datetime
    df_filtrado['data_chegada'] = pd.to_datetime(df_filtrado['data_chegada'], errors='coerce')

    # Calcular o total de pacientes por dia e hora para cada setor
    pacientes_por_dia_hora_setor = df_filtrado.groupby([df_filtrado['data_chegada'].dt.date, df_filtrado['data_chegada'].dt.hour, 'setor']).size().unstack(fill_value=0)

    # Calcular a média de pacientes por hora para cada setor
    media_por_hora_setor = pacientes_por_dia_hora_setor.groupby(level=1).mean()

    # Criar o gráfico de barras empilhadas
    ax = media_por_hora_setor.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='viridis')

    # Adicionar os valores em cada segmento da barra
    for c in ax.containers:
        labels = [f'{v.get_height():.2f}' if v.get_height() > 0 else '' for v in c]
        ax.bar_label(c, labels=labels, label_type='center')

    # Adicionar títulos e rótulos
    plt.title(f'Média de Pacientes por Hora do Dia por Especialidade - {month_name}')
    plt.xlabel('Hora do Dia')
    plt.ylabel('Média de Pacientes')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--')
    plt.legend(title='Setor')

    # Exibir o gráfico
    plt.show()
`````

# Juntando todas os meses em um único dataframe

Usando os dados dos seguintes arquivos: “dados_jan.xlsx”, “dados_fev.xlsx”, “dados_mar.xlsx”, “dados_abr.xlsx”, “dados_maio.xlsx” e “dados_jun.xlsx”, combine todos os dados mensais em um único DataFrame para realizar uma análise de tendência ao longo de todo o período, identificando padrões sazonais ou mudanças de longo prazo no volume de pacientes.

Carregue os dados de cada um dos arquivos mensais do Excel e combine-os em um único DataFrame abrangente. **Também adicionarei uma coluna chamada 'mês' para identificar a origem dos dados.**

Agora vou implementar os passos para carregar e combinar os dados dos arquivos Excel especificados. Isso envolve **criar uma lista com os caminhos dos arquivos, iterar sobre eles para carregar cada arquivo em um DataFrame, adicionar uma coluna 'mês' a cada DataFrame e, em seguida, concatená-los em um único DataFrame.**

`````python
# Juntando todas os meses em um único dataframe

import pandas as pd

# 1. Create a list of file paths
file_paths = ["janeiro25.xlsx", "fevereiro25.xlsx", "marco25.xlsx", "abril25.xlsx", "maio25.xlsx", "junho25.xlsx"]

# 2. Initialize an empty list to store DataFrames
all_months_data = []

# 3. Loop through each file path
for file_path in file_paths:
    # a. Load the Excel file
    df = pd.read_excel(file_path, usecols=[0, 1, 5, 18, 20, 22, 28, 29, 30, 31, 32, 33, 34, 35, 37, 44])

    # b. Extract the month name
    month_name = file_path.split('.')[0][:-2]

    # c. Add a 'month' column
    df['month'] = month_name

    # d. Append the DataFrame to the list
    all_months_data.append(df)

# 4. Combine all DataFrames
combined_df = pd.concat(all_months_data, ignore_index=True)

# 5. Display the first few rows of the combined DataFrame
display(combined_df.head())
`````
